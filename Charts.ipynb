````python
# Creating subplots with `plt.subplot`

# Create 10 (last parameter) evently spaced numbers within the range of 0 (first parameter) to 10 (second parameter)
x = np.linspace(0, 10, 10)
y1 = x
y2 = x**2
y3 = x**3

# plt.subplot(# of rows, # of columns, graph #)

# Chart 1 
plt.subplot(2, 2, 1)
plt.plot(x, y1, 'ro')
plt.title('$y_1 = x$')

# Chart 2
plt.subplot(2, 2, 3)
plt.plot(x, y2, 'g--')
plt.title('$y_2 = x^2$')

# Chart 3
plt.subplot(1, 2, 2)
plt.plot(x, y3, 'b^')
plt.title('$y_3 = x^3$')

# Space out the charts in the figure
plt.tight_layout()

plt.show()
````

`plt.subplots()` is a function that returns a tuple containing a figure and axes object(s). Thus when using `fig, ax = plt.subplots()` you unpack this tuple into the variables `fig` and `ax`.

Having `fig` is useful if you want to change figure-level attributes or save the figure as an image file later.

All ax's objects (the objects that have plotting methods), have a parent figure object anyway.

```python
fig, ax = plt.subplots()

# or

fig = plt.gcf() # gcf() returns/Gets the Current Figure
ax = fig.gca() # gca() returns/Gets the Current Axes

# is equaivalent to

fig = plt.figure()
ax = fig.add_subplot(111)
```

To set a group of subplots with grid(2,2) in one figure object.

```python
fig, [[ax1, ax2], [ax3, ax4]] = plt.subplots(nrows = 2, ncols = 2)
```

````python
# Creating subplots using `axis hangles`

# Create 10 (last parameter) evently spaced numbers within the range of 0 (first parameter) to 10 (second parameter)
n = 101
x = np.linspace(0, 10, n)
y1 = x
y2 = x**2
y3 = x**3
r = np.random.random(n)

# Create a figure handle (we put all the charts into one figure)
# figsize = (10, 8) increases the size of the figure
# default figure size is 8 x 6 (length x width)
fig = plt.figure(figsize = (10, 8))

# Create a super title for the entire figure (for all charts)
fig.suptitle('Super Title for All Charts')

# Chart 1
ax = fig.add_subplot(121)
ax.scatter(x, r, color = 'c')
ax.grid(True)
ax.set_title('ax')

# Chart 2
ax1 = fig.add_subplot(222)
ax1.plot(x, y2, color = 'r')
ax1.grid(True)
ax1.set_title('ax1')

# Chart 3
ax2 = fig.add_subplot(224)
ax2.plot(x, r, color = 'g')
ax2.grid(True)
ax2.set_title('ax2')

# Space out the charts in the figure
plt.tight_layout()

# Show figure
plt.show()
````

````python
# Creating subplots using `subplot2grid`

fig1 = plt.figure(1)
fig1.suptitle('Figure 1')

# `shape = (2, 2)` means we have a 2 by 2 set up
# `loc = (0, 0)` means that we want to place this graph to the top left location
ax1 = plt.subplot2grid(shape = (2, 2), loc = (0, 0))

# x, y coordinates, string, vertical, horizontal alignment of string
ax1.text(x = 0.5, y = 0.5, s = 'ax1', va = 'center', ha = 'center')

ax2 = plt.subplot2grid(shape = (2, 2), loc = (0, 1))
ax2.text(x = 0.5, y = 0.5, s = 'ax2', va = 'center', ha = 'center')

ax3 = plt.subplot2grid(shape = (2, 2), loc = (1, 0))
ax3.text(x = 0.5, y = 0.5, s = 'ax3', va = 'center', ha = 'center')

ax4 = plt.subplot2grid(shape = (2, 2), loc = (1, 1))
ax4.text(x = 0.5, y = 0.5, s = 'ax4', va = 'center', ha = 'center')

plt.tight_layout()

fig2 = plt.figure(2)
fig2.suptitle('Figure 2')

ax1 = plt.subplot2grid(shape = (3, 3), loc = (0, 0), rowspan = 1, colspan = 3)
ax1.text(x = 0.5, y = 0.5, s = 'ax1', va = 'center', ha = 'center')

ax2 = plt.subplot2grid(shape = (3, 3), loc = (1, 0), rowspan = 1, colspan = 2)
ax2.text(x = 0.5, y = 0.5, s = 'ax2', va = 'center', ha = 'center')

ax3 = plt.subplot2grid(shape = (3, 3), loc = (1, 2), rowspan = 2, colspan = 1)
ax3.text(x = 0.5, y = 0.5, s = 'ax3', va = 'center', ha = 'center')

ax4 = plt.subplot2grid(shape = (3, 3), loc = (2, 0), rowspan = 1, colspan = 1)
ax4.text(x = 0.5, y = 0.5, s = 'ax4', va = 'center', ha = 'center')

ax5 = plt.subplot2grid(shape = (3, 3), loc = (2, 1), rowspan = 1, colspan = 1)
ax5.text(x = 0.5, y = 0.5, s = 'ax5', va = 'center', ha = 'center')


plt.tight_layout()
plt.show()
````

````python
data = ((3, 10, 3), (10, 3, 4), (10, 3, 2), (5, 8, 1), (5, 1, 5))

# Number of items in each tuple
number_of_items = len(data[0])

# Width of each bar
w = 0.75

# Divide the width by number of items in each tuples
bar_width = w / number_of_items

# Create the list for x-axis
x = np.arange(len(data))

for i in range(len(data[0])):
    
    # List of tuple elements
    y = [d[i] for d in data]
    
    # plt.bar(x, y, width of the bars)
    plt.bar(x + (i * bar_width), y, bar_width)

# Places the x-axis ticks in the middle of two bars
# map(str, x) maps x to type string
plt.xticks(x + bar_width, map(str, x))

plt.show()
````
